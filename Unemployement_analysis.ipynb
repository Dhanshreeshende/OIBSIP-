```python
import datetime as dt
import calendar
import pandas as pd

# Assuming df is already loaded and the 'Date' column is properly formatted:
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract year and numeric month
df['Year'] = df['Date'].dt.year
df['Month_num'] = df['Date'].dt.month

# Convert numeric month to abbreviated month name, handling missing values
df['Month'] = df['Month_num'].apply(lambda x: calendar.month_abbr[int(x)] if pd.notnull(x) else None)

# Display the first few rows to verify the result
print(df[['Date', 'Year', 'Month_num', 'Month']].head())
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Plot correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Unemployement By Month
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 10))
sns.barplot(x="Month", y="Estimated Unemployment Rate (%)", hue="Year", data=df, palette="viridis")
plt.title("Unemployment in India by Month")
plt.xlabel("Month")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.show()
```

```python
#Unemployement Rate in India
import plotly.express as px

unemployment_rate = df.groupby("Region", as_index=False)["Estimated Unemployment Rate (%)"].mean()
figure = px.sunburst(unemployment_rate, path=["Region"],
                     values='Estimated Unemployment Rate (%)',
                     width=700, height=700,
                     title="Average Unemployment Rate in India")
figure.show()
```
